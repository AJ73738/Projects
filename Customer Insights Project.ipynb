```python

#importing libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')


%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #Adjusts the configuration of the plot we will create

#read in the data

df = pd.read_csv(r'C:\Users\user\Downloads\shopping_behavior_updated.csv')
```

```python
#Let's look at the data
df
```

```python
# 1. Customer Demographics
# Average age of customers
average_age = df['Age'].mean()
print(f"Average Age: {average_age:.2f}")
```

```python
# Gender distribution
gender_distribution = df['Gender'].value_counts()
print("Gender Distribution:\n", gender_distribution)
```

```python
# Location with the highest number of customers
most_popular_location = df['Location'].mode().values[0]
print(f"Most Popular Location: {most_popular_location}")
```

```python
# Visualizations
# Gender distribution plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Purchase Patterns
# Average purchase amount
average_purchase_amount = df['Purchase Amount (USD)'].mean()
print(f"Average Purchase Amount: ${average_purchase_amount:.2f}")
```

```python
# Average purchase amount by category
average_purchase_by_category = df.groupby('Category')['Purchase Amount (USD)'].mean()
print("Average Purchase Amount by Category:\n", average_purchase_by_category)
```

```python
# Correlation between review ratings and purchase amounts
correlation_review_purchase = df['Review Rating'].corr(df['Purchase Amount (USD)'])
print(f"Correlation between Review Ratings and Purchase Amounts: {correlation_review_purchase:.2f}")
```

```python
# Visualizations
# Average purchase amount by category bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Purchase Amount (USD)', data=df)
plt.title('Average Purchase Amount by Category')
plt.xlabel('Category')
plt.ylabel('Average Purchase Amount (USD)')
plt.show()
```

```python
# 3. Seasonal Trends
# Season with the highest number of purchases
most_popular_season = df['Season'].mode().values[0]
print(f"Most Popular Season: {most_popular_season}")
```

```python

# Visualizations
# Seasonal distribution plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Season', data=df)
plt.title('Seasonal Distribution of Purchases')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

```python
# 4. Subscription and Shipping Analysis
# Percentage of customers subscribed
subscription_percentage = df['Subscription Status'].value_counts(normalize=True) * 100
print("Subscription Percentage:\n", subscription_percentage)
```

```python
# Most commonly chosen shipping type
most_common_shipping = df['Shipping Type'].mode().values[0]
print(f"Most Common Shipping Type: {most_common_shipping}")
```

```python
# Percentage of customers applying discounts
discount_percentage = df['Discount Applied'].value_counts(normalize=True) * 100
print("Discount Percentage:\n", discount_percentage)
```

```python
# Percentage of orders with promo codes
promo_code_percentage = df['Promo Code Used'].value_counts(normalize=True) * 100
print("Promo Code Usage Percentage:\n", promo_code_percentage)
```

```python
# Visualizations
# Promo Code usage pie chart
plt.figure(figsize=(8, 8))
plt.pie(promo_code_percentage, labels=promo_code_percentage.index, autopct='%1.1f%%', startangle=90)
plt.title('Promo Code Usage Percentage')
plt.show()
```

```python
# 5. Payment Methods
# Most popular payment methods
popular_payment_methods = df['Payment Method'].value_counts()
print("Popular Payment Methods:\n", popular_payment_methods)
```

```python
# Visualizations
# Payment method distribution plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Payment Method', data=df)
plt.title('Payment Method Distribution')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
# 6. Customer Loyalty
# Correlation between previous purchases and purchase amounts
correlation_previous_purchases = df['Previous Purchases'].corr(df['Purchase Amount (USD)'])
print(f"Correlation between Previous Purchases and Purchase Amounts: {correlation_previous_purchases:.2f}")
```

```python
# Visualizations
# Scatter plot for Previous Purchases vs Purchase Amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Previous Purchases', y='Purchase Amount (USD)', data=df)
plt.title('Correlation between Previous Purchases and Purchase Amounts')
plt.xlabel('Previous Purchases')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python

# 7. Size and Color Preferences
# Most common sizes and colors
common_sizes = df['Size'].value_counts()
common_colors = df['Color'].value_counts()
print("Most Common Sizes:\n", common_sizes)
print("Most Common Colors:\n", common_colors)
```

```python
# Visualizations
# Bar plot for Most Common Sizes
plt.figure(figsize=(10, 6))
sns.barplot(x=common_sizes.index, y=common_sizes.values)
plt.title('Most Common Sizes')
plt.xlabel('Size')
plt.ylabel('Count')
plt.show()
```

```python
# 8. Frequency of Purchases
# Average purchase frequency
average_purchase_frequency = df['Frequency of Purchases'].value_counts()
print("Average Purchase Frequency:\n", average_purchase_frequency)
```

```python
# Visualizations
# Bar plot for Average Purchase Frequency
plt.figure(figsize=(10, 6))
sns.barplot(x=average_purchase_frequency.index, y=average_purchase_frequency.values)
plt.title('Average Purchase Frequency')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Count')
plt.show()
```
