```python

#importing libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')


%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #Adjusts the configuration of the plot we will create

#read in the data

df = pd.read_csv(r'C:\Users\user\Desktop\survey_results_public.csv')
```

```python
#Let's look at the data
df
```

```python
#Displaying all columns for data cleaning

pd.set_option('display.max_columns',None)
```

```python
#droping columns
df = df.drop(columns = ["Sexuality","Trans","Knowledge_1","Knowledge_2","Knowledge_3","Knowledge_4","Knowledge_5","Knowledge_6","Knowledge_7","TrueFalse_1","TrueFalse_2","TrueFalse_3"])
df
```

```python
#removing null values
df.fillna('')
```

```python
#exploratory data analysis 1
#Do most developers work remotely?
df.RemoteWork.value_counts()
```

```python
eda1 = df.RemoteWork.value_counts()

eda1.plot(kind = "bar")

#yes most developers do work remotely
```

```python
#exploratory data analysis 2
#How does coding experience affect the level of pay?
df.groupby("YearsCode")["CompTotal"].min().reset_index().plot(kind = "line")
#The higher the experience the higher the minimum salary
```

```python
#exploratory data analysis 3
#What’s the most popular method of learning to code?
df.LearnCode.value_counts()
#School is the most popular method of learning to code
```

```python
#exploratory data analysis 4
#Are you more likely to get a job as a developer if you have a Master’s degree?
df.groupby('EdLevel')['Employment'].count()
```

```python
eda4 = df.groupby('EdLevel')['Employment'].count()

eda4.plot(kind = "bar")
#No you arec not more likely to get a job as an MSc holder
```
